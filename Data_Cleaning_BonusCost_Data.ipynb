```python
import numpy as np
import pandas as pd
```

```python
bonus_cost = pd.read_excel('BonusCost_Data.xlsx', header=1)
```

```python
#removing 'Unnamed: 0' column permanently
bonus_cost.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
#starting indexing from 1
bonus_cost.index+=1
bonus_cost
```

```python
#renaming the column's name
bonus_cost.rename({'Src_PLAYER_ID': 'Player_Id', 'ActivityMonth':'Date',
                   'Product': 'Product', 'BONUS_COST': 'Bonus_Cost'}, axis=1, inplace=True)
```

```python
bonus_cost['Date'] = bonus_cost['Date'].dt.strftime('%d/%m/%Y')
```
